```python
#abrir archivo 2
df2 = pd.read_csv("world-data-2023_part2.csv",index_col=0)
```

```python
# Verificamos los 5 primeros
df2.head()
```

```python
#combinar datos de ambos dataframe 

"""se utilizpo este método ya que es el que nos permite fusionar ambos dataframe"""

df_unido= df1.merge(df2,left_on="Country",right_on="country")

df_unido.head()
```

```python
#sacamos una copia
df_unido_2=df_unido.copy()
```

```python
# 1 Después de la unión de datos, tenemos dos columnas de "country". Elimina una de ellas.

df_unido_2.drop("country",axis=1,inplace=True)

```
