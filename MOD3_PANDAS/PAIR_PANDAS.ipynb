```python
#¿Qué país consiguió más de 200 medallas de oro?
df_medallas.loc[df_medallas.Oro > 200]
```

```python
#¿Qué país consiguió más de 200 medallas de oro y más de 1000 de plata?

resultado_1 = df_medallas[(df_medallas["Oro"] > 200) & (df_medallas["Plata"] > 1000)]
print(resultado_1["País"])
```

```python
#¿Qué país consiguió más de 200 medallas de oro o más de 700 de bronce?

resultado_2 = df_medallas[(df_medallas["Oro"] > 200) | (df_medallas["Bronce"] > 700)]
print(resultado_2["País"])
```

- Momento de trabajar con Series. Contesta a las siguientes preguntas:

Crea dos listas:

```python
#Una que se llame "valores" que tenga 7 números aleatorios.

lista = list(np.random.rand(7))
# lista es un array de una dimensión, por eso se ha convertido a lista para que sea más fácil de leer
lista
```

```python
#Otra que se llame "indices" que tenga 7 strings que se llamen como los días de la semana.

indices = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]
#indices es una lista de strings que se ha creado para ser utilizada como índices en el DataFrame
indices
```

```python
#A partir de estas dos listas crea una Serie donde la variable "valores" sean los valores de la Serie y la variable indices sean los índices 
# de la Serie.

serie = pd.Series(data=lista, index=indices)
serie
```
